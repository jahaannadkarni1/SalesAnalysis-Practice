##### What time should we display ads to maximize likelihood of customer buying product

```python
# As there is a chance of the order date format changing and grabbing the
# time doesn't seem like a good idea, we instead convert the time into
# a datetime object
#the datetime object is smart enough to understand the format of the date
#and time hence we don't need to worry about the change in the format
```

```python
#converting the column to a datetime column using the to_datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
#this creates a table of the hour column which can be used to form a 
# graph of the frequency of all the times the order was placed
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
```

```python
hours = [hour for hour,df in df.groupby('Hour')] #important parameter

plt.plot(hours,df.groupby(['Hour']).count(),'b.-')

plt.grid()

plt.xticks(hours)

plt.title('Time of Order Placed')

plt.xlabel('Time in 24hr')

plt.ylabel('Frequency')

plt.show()
```

```python
#here what we did is basically make a new variable to call the dataframe
# then we filter out the values in the present df which dont have a duplicate
# value. in this keep parameter ensures all the duplicated values remain in
#in the new df from which we can see a trend
dftemp = df[df['Order ID'].duplicated(keep=False)]
```

```python
#here we created a new column called grouped in which we performed a groupby 
#with Order ID indexing the product. Here transform is basically to perform
#operations like adding subtraction but instead here we have joined the 
#products with similar order id. 
#first i had changed name but forgot to use it
#now i will change the variables again and perform the transform
dftemp['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:
                                                            ",".join(x))


    
    
```

```python
#so here we create a new variable called df1, where we clubbed the order id
#grouped then we dropped the duplicated as we have a rough idea with one 
#order id that which product was bought along with the 
df1 = dftemp[['Order ID','Grouped']].drop_duplicates()
```

```python
df1
```

```python
#now we need to iterate through the grouped column to find the pairs of 
#items that are most bought in pairs
#we use itertools and collections to count the pairs of items

from itertools import combinations
from collections import Counter
```

```python
#we initiate with the count variable which will be the counter counting

count = Counter()

#here we have a for loop to iterate through the column
for row in df1['Grouped']:
    row_list = row.split(',') #this splits the row into a list, LIST IS IMP
                            #because there we had a list hence we need one here
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key,value)
```
